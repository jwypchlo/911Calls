```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
#Top 5 zipcodes
df['zip'].value_counts().head(5)
```

```python
#Top 5 townships
df['twp'].value_counts().head(5)
```

```python
# Unique title codes
df['title'].nunique()
```

```python
# New column 'Reason'
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(df['Reason'],palette='viridis')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
# converting data type to DateTime objects
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
# creating 3 new columns
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')

# relocating legend
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# relocating legend
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0.)
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda time: time.date())
```

```python
df.head()
```

```python
df.groupby('Date').count()['twp'].plot(figsize=(10,5))
```

```python
df.groupby('Date').count()['twp'].plot(figsize=(10,5))
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot(figsize=(10,5))
plt.title('Traffic')
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot(figsize=(10,5))
plt.title('Fire')
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot(figsize=(10,5))
plt.title('EMS')
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
sns.clustermap(dayHour,cmap='viridis')
```
